```python
# Render our plots inline
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.style.use('ggplot')  # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (15, 5)
```

# Reading data from a csv file

```python
grafitti_data = pd.read_csv('Graffiti_30_Days.csv', parse_dates=['OPEN DATE'], index_col=['OPEN DATE'])

# Sampling data set
grafitti_data[:3]
```

```python
grafitti_data.loc[:, 'DAY'] = grafitti_data.index.day
grafitti_data.loc[:, 'MONTH'] = grafitti_data.index.month
grafitti_data.loc[:, 'YEAR'] = grafitti_data.index.year
grafitti_data.loc[:, 'DATE'] = pd.to_datetime(grafitti_data.YEAR*10000+grafitti_data.MONTH*100+grafitti_data.DAY,format='%Y%m%d')
grafitti_data[:5]
```

# Formatting OPEN DATE Column

I want to change the format of the dates in OPEN DATE so that months are displayed first in the format mm/dd/yyyy. Also I dont need time so I'll get rid of that as well. <br/>

Format:
FROM: YY/MM/DD HH:MM:SS    -> TO: MM/DD/YY

```python
grafitti_data['DATE'] = grafitti_data.apply(lambda x: x.strftime('%m/%d/%Y'))
```

```python
# Lets check it
grafitti_data[:3]
```

## Statistics

#### Lets break down the graffiti data.

## By District

#### Heres a link to wikipedia with a map showing which neighborhoods each district covers.

https://en.wikipedia.org/wiki/San_Francisco_Board_of_Supervisors

```python
grafittiCounts_by_district = grafitti_data['SUPERVISOR DISTRICT'].value_counts()
grafittiCounts_by_district.plot(kind='bar')
```

### Remove District 0

#### There is no district 0 so lets remove that from our graph.

```python
grafittiCounts = grafitti_data[grafitti_data['SUPERVISOR DISTRICT'] != 0]
grafittiCounts_by_district = grafittiCounts['SUPERVISOR DISTRICT'].value_counts()
grafittiCounts_by_district.plot(kind='bar')
```

## By Type

```python
grafittiCounts_by_Type = grafitti_data['TYPE'].value_counts()
# Get the top 5 types
grafittiCounts_by_Type[:5].plot(kind='bar')
```
